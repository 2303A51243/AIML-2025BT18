```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('lending_data.csv')
```

```python
print(data.dtypes)
```

```python
print(data.columns)
```

```python
print(data['credit.policy'].value_counts())
```

```python
sns.lmplot(x='fico', y='int.rate', hue='not.fully.paid', col='credit.policy', data=data, height=6)
plt.show()
```

```python
data = pd.get_dummies(data, columns=['purpose'], drop_first=True)
```

```python
X = data.drop('not.fully.paid', axis=1)
Y = data['not.fully.paid']
```

```python
!pip install scikit-learn  # Install scikit-learn if not already installed

# Import necessary modules
import pandas as pd
from sklearn.model_selection import train_test_split

# ... (Rest of your code) ...

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
!pip install scikit-learn  # Install scikit-learn if you haven't already
from sklearn.ensemble import RandomForestClassifier # Import RandomForestClassifier from sklearn.ensemble

rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X_train, Y_train)
```

```python
predictions = rfc.predict(X_test)
```

```python
!pip install scikit-learn  # Install scikit-learn if you haven't already
from sklearn.metrics import confusion_matrix # Import confusion_matrix from sklearn.metrics
predictions = rfc.predict(X_test)

cm = confusion_matrix(Y_test, predictions) # Now confusion_matrix is defined and accessible
print("Confusion Matrix:")
print(cm)
```

```python
!pip install scikit-learn  # Install scikit-learn if you haven't already
from sklearn.metrics import classification_report # Import classification_report from sklearn.metrics

cr = classification_report(Y_test, predictions)
print("Classification Report:")
print(cr)
```
